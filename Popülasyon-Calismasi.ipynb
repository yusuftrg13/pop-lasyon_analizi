```python
#Kütüphaneyi Ekleyelim
import pandas as pd
```

```python
#Veri Setini Yükle
df=pd.read_csv('Desktop\popülasyon.csv')
```

```python
#Sayısal sütunlar icin özet istatiksel alalım
df.describe()
```

```python
#Veri Temizleme Yapalım
df.isnull().sum()
```

```python
#Nüfusun Yıllık Değişimi

df['Nüfus Degisimi']=df['2023']-df['2022']
print(df['Nüfus Degisimi'])
```

```python
max_sehir=df.loc[df['2023'].idxmax()]
min_sehir=df.loc[df['2023'].idxmax()]

print(max_sehir)
print(min_sehir)
```

```python
#Toplam Nüfus Artışı
df['Toplam Artış']=df['2023']-df['2007']
print(df['Toplam Artış'])
```

```python
#Her Şehir icin Ortalama Nüfus
df['Ortalama Nüfus']=df.loc[:,'2007':'2023'].mean(axis=1)
print(df['Ortalama Nüfus'])
```

```python
#Tüm şehirler icin yıllara göre nüfus orqnı artış çizelgesi
df.loc[: , '2007':'2023'].sum().plot(kind='line')
```

```python
import matplotlib.pyplot as plt

# 'il' sütununu index olarak ayarlayıp 2023 yılına ait nüfus verilerini bar grafiği olarak çizelim
df.set_index('il')['2023'].plot(kind='bar', figsize=(12, 6))

# Grafiği düzenleyelim
plt.title("2023 Yılı Şehir Bazında Nüfus Dağılımı", fontsize=16)
plt.xlabel("Şehirler", fontsize=12)
plt.ylabel("Nüfus", fontsize=12)
plt.xticks(rotation=90, fontsize=10)  # Şehir isimlerinin dikey görünmesi için döndürelim
plt.tight_layout()  # Grafik elemanlarının sıkışmasını önleyelim
plt.grid(axis='y')  # Y eksenine grid ekleyelim

# Grafiği gösterelim
plt.show()
```

```python
korelasyon_analizi=df.loc[:,'2007':'2023'].corr()
print(korelasyon_analizi)
```
